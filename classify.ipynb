```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import normalize
from sklearn.cluster import AgglomerativeClustering
import re
```

```python
delim = "--------------------------------------------------------------------------------------------------"
```

```python
def count_error(path):
    with open(path) as file:
        ar = file.read()
        count = re.findall(rf'{delim}\n', ar)
        return len(count)
```

```python
p = ["dataset/error logs.txt","dataset/error_log.txt","dataset/error_log1.txt","dataset/ERROR-LOG.txt","dataset/streamliterrorlog.txt"]
```

```python
e = []
for i in p:
    e.append(count_error(i))
e
```

```python
a = ""
for path in p:
    with open(path) as file:
        ar = file.read()
        a += ar
```

```python
df = pd.DataFrame(a.split(delim),columns=["error"])
df
```

```python
df = df.replace(to_replace ='\n', value = '', regex = True)
df
```

```python
df = df[df["error"] != ""]
df
```

```python
df = df[df["error"].str.contains("-----")==False]
df
```
